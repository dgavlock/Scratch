```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
import scipy.stats
plt.style.use('seaborn')
```

```python
def U(x):
    Eb = 5 #kbt
    d = 1 # Agnstrom
    return Eb*((x/d)**2-1)**2
```

```python
x = np.linspace(0,100, 100)

y = U(x)

plt.plot(x, y)
```

```python
def dampen(xj, dt):
    m = 10**-9
    gamma = 10**9
    Eb = 5
    Uderivative = sym.diff(U(xj))
    delta_x_rand = np.random.choice([-np.sqrt(2*dt/(m*gamma)), np.sqrt(2*dt/(m*gamma))])
    return xj - (dt/m*gamma) * Uderivative + delta_x_rand
    
```

```python
xj = -1
steps = 100
dt = 0.001

values = [U(xj)]

for i in range(steps-1):
    xj = dampen(xj, dt)
    potential = U(xj)
    values.append(potential)

values2 = [xj]
    
for i in range(steps-1):
    xj = dampen(xj, dt)
    values2.append(xj)
```

```python
plt.plot(range(steps), values2, '-r')
```

```python
plt.plot(range(steps), values, '-')
```
